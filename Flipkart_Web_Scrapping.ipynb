### Fetching Real Time SmartPhones Details From FlipKart Website Using Web Scraping

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

df= pd.DataFrame()
for j in range(1,21):
    curr_page = j
    product = 'smart phones'
    url = "https://www.flipkart.com/search?q="
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.101 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9'
    }
    query_url= url + '+'.join(product.split()) + '&page={}'.format(curr_page)
    r = requests.get(query_url,headers)
    text = r.text
    soup = BeautifulSoup(text,'html.parser')
    result_page = soup.find_all('div',{'class':'_1YokD2 _3Mn1Gg'})[1]
    titles = [i.text for i in result_page.find_all('div',{'class':'_4rR01T'})]
    #original_price = [int(('').join(i.text[1:].split(',')))  for i in result_page.find_all('div',{'class':'_3I9_wc _27UcVY'})]
    disc_price = [int(('').join(i.text[1:].split(','))) for i in result_page.find_all('div',{'class':'_30jeq3 _1_WHN1'})]
    avg_rating = [i.text for i in result_page.find_all('div',{'class':'_3LWZlK'})]
    total_ratings = [int(''.join(i.text.split('\xa0&\xa0')[0].split(' ')[0].split(','))) for i in result_page.find_all('span',{'class':'_2_R_DZ'})]
    reviews = [int(''.join(i.text.split('\xa0&\xa0')[1].split(' ')[0].split(','))) for i in result_page.find_all('span',{'class':'_2_R_DZ'})]
    RAM_Size = [i.text.split('|')[0][0:-4] for i in result_page.find_all('div',{'class':'fMghEO'})]
    ROM_Size = [i.text.split('|')[1][0:-4] for i in result_page.find_all('div',{'class':'fMghEO'})]
    df = pd.concat([df,pd.DataFrame({'Title':titles,'Discount Price':disc_price,'Avg. Rating':avg_rating,'Total Ratings':total_ratings,'Reviews':reviews,'RAM Size':RAM_Size,'ROM Size':ROM_Size})])
    df.index = [i for i in range(0,len(df))]
    df['ROM Size'] = df['ROM Size'].apply(lambda x: x[1:7] if len(x)>6 else x)
    df.drop(df[(df['ROM Size']=='12MP F') | (df['ROM Size']=='7MP Fr')].index,axis=0,inplace=True)
```
